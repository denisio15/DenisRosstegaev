```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="ticks")
```

```python
data = pd.read_csv('happyscore_income.csv', sep=",")
```

```python
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
data.describe()
```

```python
mean_happy = data['happyScore'].mean()
median_happy = data['happyScore'].median()
data['Happy_binary'] = (data['happyScore'] > mean_happy).astype(int)
data['Happy_binary'].unique()
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='adjusted_satisfaction', y='avg_satisfaction', data=data)
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.scatterplot(ax=ax, x='adjusted_satisfaction', y='avg_satisfaction', data=data, hue='Happy_binary')
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data['avg_income'])
```
